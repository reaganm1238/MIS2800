```python
%%sql
CREATE TABLE IF NOT EXISTS stock_performance (
    ticker TEXT,
    price_before_game REAL,
    price_week_after_game REAL,
    trading_volume_millions INTEGER
) STRICT;
```

```python
%%sql
CREATE Ad_Buys (
    ticker TEXT PRIMARY KEY,
    company_name TEXT,
    ad_cost_millions REAL,
    airtime_seconds INTEGER
) STRICT;
```

```python
# Instructions: Show all the columns from the stock_performance table
%%sql
SELECT * FROM stock_performance;
```

```python
# # Instructions: Show the ticker and price before the game
# sort by ticker in alphabetical order from the stock_performance table
%%sql
SELECT ticker, price_before_game FROM stock_performance ORDER BY ticker;
```

```python
# # Instructions: Show all the columns for the brands that have a ticker
# beginning with the letter T in the stock_performance table
%%sql
SELECT * FROM stock_performance WHERE ticker LIKE 'T%';
#
```

```python
# # Instructions: Show only the ticker and price_after_game
# from the stock_performance table if the price is above 150
%%sql
SELECT ticker, price_week_after_game FROM stock_performance WHERE price_week_after_game > 15
```

```python
# Instructions: Show all the columns from the ad_buys table
%%sql
SELECT * FROM ad_buys;
```

```python
# # Instructions: Show the ticker and company_name from the ad_buys table in
# sort in descending order by company_name
%%sql
SELECT ticker, company_name FROM ad_buys ORDER BY company_name DESC;
```

```python
# # Instructions: Show the ticker and airtime seconds from the ad_buys table
# Filter to show only if the ticker begins with T and and airtime seconds is 0
%%sql
SELECT ticker, airtime_seconds FROM ad_buys WHERE ticker LIKE 'T%' AND airtime_seconds = 0
```
