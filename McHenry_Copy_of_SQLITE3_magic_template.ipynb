```python
%%sql /* 1. Create the table with STRICT types */ CREATE TABLE products ( product_id INTEGER PRIMARY KEY, name TEXT, price REAL ) STRICT; /* 2. Insert at least 3 items (Perform a 'Create' operation) */

INSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00);
```

```python
%%sql /* Show all products that cost more than $50
SELECT from products WHERE price > 50
```

```python
%%sql /* Update a price */
UPDATE products SET price = 99.99 WHERE name = 'Superbowl Jersey'; /* Delete an item */ DELETE FROM products WHERE product_id = 1;
```
